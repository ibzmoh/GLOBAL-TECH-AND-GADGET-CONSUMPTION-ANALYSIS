# GLOBAL TECH AND GADGET CONSUMPTION ANALYSIS:

## 1. Import packages

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## 2. Data Loading and Inspection

```python
tech_data = pd.read_csv('Global_Tech_Gadget_Consumption.csv')
print(tech_data.head())
tech_data.groupby('Country')[['Smartwatch Penetration (%)', '5G Penetration Rate (%)']].describe()
```

The mean Smartwatch Penetration (%) for all countries ranges between 12% - 16%, with Brazil having the highest rate at 16.03% and China having the lowest at 12.09%. The UK and USA follow closely behind Brazil with rates of 15.87% and 15.58% respectively. South Korea and India show moderate penetration at 14.64% and 14.20%, while Canada, Japan, Germany, and France fall between 12.87% and 13.60%.

The mean 5G Penetration Rate (%) shows greater variation across countries, with France leading at 58.00%, followed by the USA at 54.27%. Both countries significantly outpace others in 5G adoption. Germany, the UK, Japan, and South Korea form a middle group with rates between 42% and 44%. India (37.25%), Canada (33.21%), and China (30.77%) show more modest 5G penetration, while Brazil has the lowest rate at 41.99%. Notably, the maximum 5G penetration reaches above 70% in several countries, with France and South Korea showing the highest peaks at approximately 79%.

```python
tech_data.groupby('Country')[['Smartphone Sales (Millions)', 'Laptop Shipments (Millions)']].describe()
```

The mean Smartphone Sales (Millions) for all countries is in the 200 million range, with the US having the lowest smartphone sales of the 10 countries with a mean value of 223 million sales over the 11 years. The UK has the highest mean value of 330 million smartphone sales in the last 11 years.

For Laptop Shipments (Millions), Canada, China and Brazil have the highest mean values of 90+ million laptops shipped in the last 11 years. The US, UK and Japan have the lowest laptop shipments with values of 77, 80 and 81 million respectively. Despite the UK having the highest mean Smartphone Sales, they have one of the smallest amounts of laptops shipped with an average of 80 million laptops shipped per year.

```python
usa_data = tech_data[tech_data.Country == 'USA']
China_data = tech_data[tech_data.Country == 'China']
India_data = tech_data[tech_data.Country == 'India']
Germany_data = tech_data[tech_data.Country == 'Germany']
Uk_data = tech_data[tech_data.Country == 'UK']
france_data = tech_data[tech_data.Country == 'France']
Japan_data = tech_data[tech_data.Country == 'Japan']
South_Korea_data = tech_data[tech_data.Country == 'South Korea']
Brazil_data = tech_data[tech_data.Country == 'Brazil']
Canada_data = tech_data[tech_data.Country == 'Canada']
```
